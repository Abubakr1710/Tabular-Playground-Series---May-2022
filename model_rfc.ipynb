```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
df.head(5)
```

```python
df.drop(['id', 'f_27'], axis=1, inplace=True)
```

```python
df.head()
```

```python
X, y  = df.drop(['target'], axis=1), df['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import accuracy_score
num_vars  = ['f_00', 'f_01', 'f_02', 'f_03', 'f_04', 'f_05', 'f_06','f_19', 'f_20', 'f_21', 'f_22', 'f_23', 'f_24', 'f_25','f_26', 'f_28']
num_prepro = Pipeline(steps=[('scaler', StandardScaler())])
tree_pre  = ColumnTransformer(transformers=[('num', num_prepro, num_vars)], remainder='passthrough')

pipe = Pipeline([('tree_preprosessor', tree_pre), ('dtc', DecisionTreeClassifier())])
```

```python
pipe.fit(X_train, y_train)
```

```python
pred = pipe.predict(X_test)
```

```python
acc  = accuracy_score(y_test, pred)
```
